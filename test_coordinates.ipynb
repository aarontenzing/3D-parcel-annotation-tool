```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('wireframes\img1_1.png')
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
import json

f = open('annotations.json')
data = json.load(f)

for i in data:
    if (i['image'] == 1 and i['shot'] == 1):
        points = i['projection']
    else: 
        continue
print(points)

edges = (
            (0, 1),
            (1, 2),
            (2, 3),
            (3, 0),
            (4, 5),
            (5, 6),
            (6, 7),
            (7, 4),
            (0, 4),
            (1, 5),
            (2, 6),
            (3, 7)
        )
```

```python
def drawline(image, points):
    img_copy = image.copy()
    for i in edges:
        x = points[i[0]]
        y = points[i[1]]
        cv2.line(img_copy, (x[0], 512 - x[1]) , (y[0],512 - y[1]), (255,0,0), 3)
    plt.imshow(img_copy)
```

```python
def drawpoint(image, points):
    img_copy = image.copy()
    for i in points:
        cv2.circle(img_copy, i, 0, (255,0,0), 3)
    plt.imshow(img_copy)
```

```python
drawpoint(image, points)
```
