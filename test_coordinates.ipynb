### Visualise the image

```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('wireframes\\img2_2.png')
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

### Read the coordinates from the json file

```python
import json

f = open('annotations.json')
data = json.load(f)

for i in data:
    if (i['image'] == 2 and i['shot'] == 2):
        points = i['projection']
        center = i['center']
    else: 
        continue
print(points)

edges = (
            (0, 1),
            (1, 2),
            (2, 3),
            (3, 0),
            (4, 5),
            (5, 6),
            (6, 7),
            (7, 4),
            (0, 4),
            (1, 5),
            (2, 6),
            (3, 7)
        )
```

```python
def drawline(image, points):
    img_copy = image.copy()
    for i in edges:
        x = points[i[0]]
        y = points[i[1]]
        cv2.line(img_copy, (x[0], x[1]) , (y[0], y[1]), (255,0,0), 4)
    plt.imshow(img_copy)
```

```python

def drawpoint(image, points, center):
    img_copy = image.copy()
    for i in points:
        cv2.circle(img_copy, i, 0, (255,0,0), 8)
    
    cv2.circle(img_copy, center[1], 0, (255,255,0), 8)
    plt.imshow(img_copy)
```

```python
# Drawing the vertices and center of the box
drawpoint(image, points, center)
```

```python
# drawing all the lines
drawline(image, points)
```
