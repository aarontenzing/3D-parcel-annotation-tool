### Read the coordinates from the json file

```python
import json

f = open('..\\annotations.json')
data = json.load(f)
points = data[0]["projection"]

edges = (
            (0, 1),
            (1, 2),
            (2, 3),
            (3, 0),
            (4, 5),
            (5, 6),
            (6, 7),
            (7, 4),
            (0, 4),
            (1, 5),
            (2, 6),
            (3, 7)
        )
```

```python
def drawpoint(image, points):
    img_copy = image.copy()
    for i in points[:8]:
        cv2.circle(img_copy, i, 0, (255,0,0), 5)
    
    cv2.circle(img_copy, points[8], 0, (0,255,0), 5)
    plt.imshow(img_copy)
```

```python
def drawline(image, points):
    img_copy = image.copy()
    for i in edges:
        x = points[i[0]]
        y = points[i[1]]
        cv2.line(img_copy, (x[0], x[1]) , (y[0], y[1]), (255,0,0), 2)
    plt.imshow(img_copy)
```

```python
# Drawing the vertices and center of the box
drawpoint(image, points)
```

```python
# drawing all the lines
drawline(image, points)
```

```python
def dimension(vertices):
    # Extract x, y, and z values from each vertex
    x_values = [vertex[0] for vertex in vertices]
    y_values = [vertex[1] for vertex in vertices]
    z_values = [vertex[2] for vertex in vertices]
    print()
    print("x_values: ", x_values)
    print("y_values: ", y_values)
    print("z_values: ", z_values)

    # Calculate dimensions along each axis
    width_x = abs(max(x_values)) + abs(min(x_values))
    height_y = abs(max(y_values)) - min(y_values)
    depth_z = max(z_values) - min(z_values)
    
    return width_x/2, height_y/2, depth_z/2
```
