```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

# Testing

I first started experimenting using *wireframes* as input for ControlNet (Stable Diffusion). 
I noticed when using *wireframes* with *all ribs visible*, the model couldn't interpret the depth and would draw the box in a really weird way. 

```python
# create figure 
fig = plt.figure(figsize=(10, 7)) 

rect = cv2.imread('..\\images_sd\\rect.png')
rect = cv2.cvtColor(rect, cv2.COLOR_BGR2RGB)
image1 = cv2.imread('..\\images_sd\\im1.png')
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.imread('..\\images_sd\\im2.png')
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

rows = 1
columns = 3

# Adds a subplot at the 1st position 
fig.add_subplot(rows, columns, 1) 

# showing image 
plt.imshow(rect) 
plt.axis('off') 
plt.title("input") 

fig.add_subplot(rows, columns, 2) 

# showing image 
plt.imshow(image1) 
plt.axis('off') 
plt.title("ouput 1") 

fig.add_subplot(rows, columns, 3) 
  
# showing image 
plt.imshow(image2) 
plt.axis('off') 
plt.title("output 2") 
```

Next I tried using a solid cube, including the depth map preprocessing option in ControlNet.
The issue that arises when using this strategy, the background isn't realistic. 
This happens because the depth map estimation can't be measured behind the box. 
But positivly the box geometry is much more realistic. The prompt I used for generating these pictures : "Parcel delivered by Amazon".

```python
# create figure 
fig = plt.figure(figsize=(10, 7)) 

rect = cv2.imread('..\\images_sd\\depth_input.JPG')
rect = cv2.cvtColor(rect, cv2.COLOR_BGR2RGB)
image1 = cv2.imread('..\\images_sd\\depth2.png')
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.imread('..\\images_sd\\depth1.png')
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

rows = 1
columns = 3

# Adds a subplot at the 1st position 
fig.add_subplot(rows, columns, 1) 

# showing image 
plt.imshow(rect) 
plt.axis('off') 
plt.title("input") 

fig.add_subplot(rows, columns, 2) 

# showing image 
plt.imshow(image1) 
plt.axis('off') 
plt.title("ouput 1") 

fig.add_subplot(rows, columns, 3) 
  
# showing image 
plt.imshow(image2) 
plt.axis('off') 
plt.title("output 2") 
```

So we should not use a ControlNet model that uses a depth map, because this gives SD no freedom in generating a realistic background.
I tried using Canny edge and scribble ControlNet model again, but now I used a wireframe input where not all the ribs were visible.
This gave better and more realistic results. 

```python
# create figure 
fig = plt.figure(figsize=(10, 7)) 

rect = cv2.imread('..\\images_sd\\input_canny.png')
rect = cv2.cvtColor(rect, cv2.COLOR_BGR2RGB)
image1 = cv2.imread('..\\images_sd\\canny1.png')
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.imread('..\\images_sd\\canny2.png')
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

rows = 1
columns = 3

# Adds a subplot at the 1st position 
fig.add_subplot(rows, columns, 1) 

# showing image 
plt.imshow(rect) 
plt.axis('off') 
plt.title("input") 

fig.add_subplot(rows, columns, 2) 

# showing image 
plt.imshow(image1) 
plt.axis('off') 
plt.title("ouput 1") 

fig.add_subplot(rows, columns, 3) 
  
# showing image 
plt.imshow(image2) 
plt.axis('off') 
plt.title("output 2") 
```

- Stable diffusion model : checkpoint dreamlike-photoreal-2.0.ckpt
- ControlNet model: control_v11p_sd15_canny 
- prompt: "amazon parcel that is lying on a shelf in the warehouse"
- preprocessor: none

### Binary mask of picture and applied to Stable diffusion generated picture.

```python
# create figure 
fig = plt.figure(figsize=(10, 7)) 

mask = cv2.imread('..\\images_sd\\input_canny_mask.png')
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)
#mask = np.invert(mask)

image = cv2.imread('..\\images_sd\\canny2.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image_mask = image * mask

rows = 1
columns = 3

# Adds a subplot at the 1st position 
fig.add_subplot(rows, columns, 1) 

# showing image 
plt.imshow(mask) 
plt.axis('off') 
plt.title("binary mask") 

fig.add_subplot(rows, columns, 2) 

# showing image 
plt.imshow(image) 
plt.axis('off') 
plt.title("original") 

fig.add_subplot(rows, columns, 3) 

# showing image 
plt.imshow(image_mask) 
plt.axis('off') 
plt.title("mask applied") 
```
