```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('wireframes\img1_1.png')
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
import json

f = open('annotations.json')

data = json.load(f)

for i in data:
    if (i['image'] == 1 and i['shot'] == 1):
        points = i['projection']
    else: 
        continue
    
print(points)
```

```python
img_copy = image.copy()
for i in points:
    cv2.circle(img_copy, i, 0, (255,0,0), 3)
plt.imshow(img_copy)
```
